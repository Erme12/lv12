**How do you think about a deferred disability annuity**

- Once you get disabled you get the PV of the disability annuity as a lump sum
- $C_{x+k}^{diabled} = \ddot{a}_{x+k}^{diab}$

```python
def aexdis(gender, age, iRate=0.02, bChatter = False):
    omega = 110
    Vx = 0
    for i in range(omega-1,age-1,-1):
        if i < 65:
            px = 1 - Evk90(gender,i,"QXI")
        else:
            px = 1 - Evk90(gender,i,"QX")
        Vx = 1 + px / (1+iRate) * Vx
    if (bChatter): print(Vx)
    return(Vx)

print("Example:",aexdis(0, 30, iRate=0.02))

def defdis(gender, age, s=65, iRate=0.02):
    Vx = 0 
    for i in range(s-1,age-1,-1):
        ix = Evk90(gender,i,"XI")
        px = 1  - Evk90(gender,i,"QXA")
        Vx = ix / (1+iRate) * aexdis(gender, i, iRate=iRate) + px / (1+iRate) * Vx
    return(Vx)
        
print(defdis(0, 30, s=65, iRate=0.02))

x = range(20,65)
ym = []
yf = []

for i in x:
    ym.append(defdis(0, i, s=65, iRate=0.02))
    yf.append(defdis(1, i, s=65, iRate=0.02))

plt.figure(1)
plt.plot(x,ym,"-r",x,yf,"gp")
plt.grid(True)


x = range(20,65)
ym = []
yf = []

for i in x:
    ym.append(aexdis(0, i, iRate=0.02))
    yf.append(aexdis(1, i, iRate=0.02))

plt.figure(2)
plt.plot(x,ym,"-r",x,yf,"gp")
plt.grid(True)
```

```python
def DisabilityAnnuity(gender,age,s=65,iRate=0.02):
    aei = 0. # \ddot{a}_x^{i}
    aea_to_i = 0. # \ddot{a}_x^{ai}
    v = 1. /(1.+iRate)
    for i in range(s-1,age-1,-1):
        ix = Evk90(gender,i,"XI")
        qxa = Evk90(gender,i,"QXA")
        qxi = Evk90(gender,i,"QXI")
        aea_to_i = v * (ix * aei + aea_to_i * (1-qxa-ix))
        aei = 1. +  v * (1-qxi) * aei
        sar = aei - aea_to_i
    return(aea_to_i,aei,sar)

x=[]
y1=[]
y2=[]
y3=[]
s=65
gender=0
for i in range(20,s):
    x.append(i)
    a,b,c = DisabilityAnnuity(gender,i,s=s)
    y1.append(a)
    y2.append(b)
    y3.append(c)
    
plt.figure(4)
plt.semilogy(x,y1,x,y2)
plt.grid(True)
plt.legend(["Defferred","In Payment"])

plt.figure(2)
plt.plot(x,y1,x,y2)
plt.grid(True)
plt.legend(["Defferred","In Payment"])

plt.figure(5)
plt.plot(x,y3,"r-")
plt.grid(True)
```

```python
def widowannuity(gender,age,iRate=0.02):
    Vx=0
    v = 1. /(1.+iRate)
    for i in range(110,age-1,-1): 
        qx = Evk90(gender,i,"QX")
        Vx = 1 + v * (1-qx) * Vx
    return(Vx)
    

def CollWiddowsPension(age, gender=0,iRate=0.02,bTrace=False):
    Vx=0
    v = 1. /(1.+iRate)
    for i in range(110,age-1,-1):
        hx = Evk90(gender,i,"WX")
        qxa = Evk90(gender,i,"QX")
        yx =  int(0.45+Evk90(gender,i,"YX"))
        ay_of_x = widowannuity(1-gender,yx)
        if bTrace:
            print("Age:",i)
            print(hx, qxa, yx, ay_of_x)
            print("Vx = ",Vx)
        Vx = qxa * hx * ay_of_x * v + (1-qxa) * v * Vx 
    return(Vx)
```
